```python
spark.sql('''select *,
      upper(customer_name) as upper_customer_name,
      current_date() as processDate
from datamodeling.bronze.bronze_table''').createOrReplaceTempView('silver_source')
```

```python
%sql
select * from silver_source
```

```python
if spark.catalog.tableExists('datamodeling.silver.silver_table'):
  pass

else:
  spark.sql('''create table datamodeling.silver.silver_table
            as select * from silver_source''')
```

```python
%sql
merge into datamodeling.silver.silver_table as trg
using silver_source as src
on trg.order_id=src.order_id
when matched then update set *
when not matched then insert *
```
